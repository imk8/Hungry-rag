```python
import googlemaps
import pprint
from GooglePlaces_API_Key import get_my_key #Imports personal API key

# Define the API Key.
API_KEY = get_my_key()

# Define the Client
gmaps = googlemaps.Client(key = API_KEY)

places_result  = gmaps.places('italian food near tampines mall',region = "SG" ) #Demo query
pprint.pprint(places_result)
```
