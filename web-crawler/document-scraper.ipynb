```python
import re  # Import the re module for regular expressions

count = 1

data = {}

for query in queries:

    params = {
        "q": query,
        "hl": "en",
        "gl": "sg",
        "start": 0,
    }

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"
    }

    page_num = 0


    while True:
        page_num += 1

        html = requests.get("https://www.google.com/search", params=params, headers=headers, timeout=30)
        soup = BeautifulSoup(html.text, 'lxml')

        for result in soup.select(".tF2Cxc"):
            title = result.select_one(".DKV0Md").text
            try:
                snippet = result.select_one(".lEBKkf span").text
            except:
                snippet = None
            link_element = result.select_one(".yuRUbf a")
            if link_element:
                link = link_element["href"]
                data[link] = {
                    "title": title,
                    "snippet": snippet
                }

        next_button = soup.select_one(".d6cvqb a[id=pnnext]")
        if next_button:
            next_url = next_button["href"]
            match = re.search(r'start=(\d+)', next_url)  # Use regular expressions to extract the numeric part
            if match:
                params["start"] = int(match.group(1))
            else:
                break  # If no numeric part is found, exit the loop
        else:
            break

print(json.dumps(data, indent=2, ensure_ascii=False))
print(len(data))
```

```python
max_iter = 10
timeout = 3

for link, entry in data.items():
    print("Looking into - ", link)
    try:
        webpage = requests.get(link, timeout=timeout)
        webpagetext = BeautifulSoup(webpage.text, "html.parser")
        all_p = webpagetext.find_all('p')
        text = ""
        count = 0
        for item in all_p:
            text = text + item.get_text()
            count+=1
            if count > max_iter:
                break
        entry['doc'] = text 
    except Exception as e :
        print(f'Exception occurred: {e}')
```
